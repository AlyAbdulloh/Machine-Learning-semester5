# 1. Buatlah model klasifikasi dengan menggunakan SVM untuk data suara, voice.csv.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io

voice = pd.read_csv(io.BytesIO(uploaded['voice.csv']))  # spesifiksi encoding diperlukan karena data tidak menggunakan UTF-8
voice.head()
```

**pisahkan fitur dengan label**

```python
X = voice.iloc[:,:-1]
y = voice.iloc[:,-1] # voice['label']

print(X.shape)
print(y.shape)
```

**split data dan melakukan ekstraksi**

```python
# Split data training dan testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)

#inisiasi SVC
clf = SVC()

# melakukan Fitting model
clf.fit(X_train, y_train)

# prediksi dengan data testing
ypred = clf.predict(X_test)
```

**Evaluasi model**

```python
from sklearn.metrics import accuracy_score

# Evaluasi akurasi data
score = accuracy_score(y_test, ypred)
print(f'Hasil akurasi data : {score}')
```

```python
from sklearn.model_selection import GridSearchCV

params = {
    'kernel': ['rbf', 'linear', 'poly'],
    'C': [1, 5, 10, 50, 100],
    'gamma': [0.0001, 0.0005, 0.001, 0.005]
}

grid = GridSearchCV(clf, params, n_jobs=-1)

%time grid.fit(X_train, y_train)

print(f'Best Estimator: {grid.best_estimator_}')
print(f'Score: {grid.best_score_}')
```

# 2. Buatlah model klasfikasi Multinomial Naive Bayes dengan ketentuan,

1. Menggunakan data spam.csv
2. Fitur CountVectorizer dengan mengaktifkan stop_words
3. Evaluasi hasilnya

```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd
import io

df = pd.read_csv(io.BytesIO(uploaded['spam.csv']), encoding='latin-1')  # spesifiksi encoding diperlukan karena data tidak menggunakan UTF-8
df.head()
```

**DROP KOLOM**

```python
# Drop 3 kolom terakhir dengan fungsi iloc
df = df.drop(df.iloc[:,2:], axis=1)

# Cek data
df.head()
```

**Mengubah nama kolom v1 dan c2**




```python
new_cols = {
    'v1': 'Labels',
    'v2': 'SMS'
}

# Rename nama kolom v1 dan v2
df = df.rename(columns=new_cols)

# cek data
df.head()
```

**melakukan encoding pada**

```python
# Data untuk label
new_labels = {
    'spam': 1,
    'ham': 0
}

# Encode label
df['Labels'] = df['Labels'].map(new_labels)

# Cek data
df.head()
```

**Memisahkan fitur dengan label**

```python
X = df['SMS'].values
y = df['Labels'].values
```

**Ekstraksi fitur menggunakan fungsi CountVectorizer**

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer

# Split data training dan testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50)

# Inisiasi CountVectorizer dengan menggunakan stop_words
bow = CountVectorizer(stop_words='english')

# Fitting dan transform X_train dengan CountVectorizer
X_train = bow.fit_transform(X_train)

# Transform X_test
X_test = bow.transform(X_test)
```

**Training dan Evaluasi model**

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Inisiasi MultinomialNB
mnb = MultinomialNB()

# Fit model
mnb.fit(X_train, y_train)

# Prediksi dengan data training
y_pred_train = mnb.predict(X_train)

# Evaluasi akurasi data training
acc_train = accuracy_score(y_train, y_pred_train)

# Prediksi dengan data testing
y_pred_test = mnb.predict(X_test)

# Evaluasi akurasi data testing
acc_test = accuracy_score(y_test, y_pred_test)

# Print hasil evaluasi
print(f'Hasil akurasi data train: {acc_train}')
print(f'Hasil akurasi data test: {acc_test}')
```

# Buatlah model klasfikasi Multinomial Naive Bayes dengan ketentuan,

1. Menggunakan data spam.csv
2. Fitur TF-IDF dengan mengaktifkan stop_words
3. Evaluasi hasilnya dan bandingkan dengan hasil pada Tugas no 2.
4. Berikan kesimpulan fitur mana yang terbaik pada kasus data spam.csv

```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd
import io

df2 = pd.read_csv(io.BytesIO(uploaded['spam (1).csv']), encoding='latin-1')  # spesifiksi encoding diperlukan karena data tidak menggunakan UTF-8
df2.head()
```

**Drop Kolom**

```python
# Drop 3 kolom terakhir dengan fungsi iloc
df2 = df2.drop(df2.iloc[:,2:], axis=1)

# Cek data
df2.head()
```

**Mengubah nama kolom v1 dan c2**


```python
new_cols = {
    'v1': 'Labels',
    'v2': 'SMS'
}

# Rename nama kolom v1 dan v2
df2 = df2.rename(columns=new_cols)

# cek data
df2.head()
```

**Memisahkan fitue dengan label**

```python
X = df2['SMS'].values
y = df2['Labels'].values
```

**Ekstraksi fitur menggunakan TF-IDF**

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

# Split data training dan testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50)

# Inisiasi TF-IDF dengan menggunakan stop_words
bow = TfidfVectorizer(stop_words='english')

# Fitting dan transform X_train dengan TF-IDF
X_train = bow.fit_transform(X_train)

# Transform X_test
X_test = bow.transform(X_test)
```

**Training dan Evaluasi model**

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Inisiasi MultinomialNB
mnb = MultinomialNB()

# Fit model
mnb.fit(X_train, y_train)

# Prediksi dengan data training
y_pred_train = mnb.predict(X_train)

# Evaluasi akurasi data training
acc_train = accuracy_score(y_train, y_pred_train)

# Prediksi dengan data testing
y_pred_test = mnb.predict(X_test)

# Evaluasi akurasi data testing
acc_test = accuracy_score(y_test, y_pred_test)

# Print hasil evaluasi
print(f'Hasil akurasi data train: {acc_train}')
print(f'Hasil akurasi data test: {acc_test}')
```
